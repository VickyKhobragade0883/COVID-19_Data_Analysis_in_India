# Weekly analysis of covid 19 data across the states

```python
import requests
import json
from pprint import pprint
import pandas as pd
from datetime import datetime,date
```

# IMPORTING DATA FROM JSON FORMAT


```python
response2= requests.get("https://data.covid19india.org/v4/min/timeseries.min.json")
results2= json.loads(response2.text)
pd.DataFrame(results2)
```

# ANALYSIS STATE BY STATE

```python
dfan=pd.DataFrame(results2['AN']['dates'])
dfan
```

```python
dfan.drop('delta',inplace =True)
```

```python
dfan.drop('delta7',inplace =True)
```

# ANALYSIS OF 'AN'

```python
df2an=dfan.transpose()
df2an
```

```python
AN=df2an['total'].apply(pd.Series)
AN

```

```python
AN.reset_index(level=0,inplace=True)
```

```python
AN.rename({'index':'dates'},axis=1,inplace=True)
```

```python
AN
```

# IMPORTING PYTHON DATA TO SQL DATABASE

```python
import mysql.connector
from mysql.connector import Error
from sqlalchemy import create_engine
```

```python
host_name = 'localhost'
db_name = 'pandas_db'
u_name = 'root'
u_pass = '12345'
port_num = '3306'
```

```python
my_eng = create_engine('mysql+mysqlconnector://' +u_name+ ':' + u_pass + '@' + host_name + ':' + port_num + 
                      '/' + db_name , echo= False)
```

```python
AN.to_sql(name='AN', con=my_eng, if_exists='append', index=False)
```
